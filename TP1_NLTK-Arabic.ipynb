```python
import nltk
import string
```

```python
Ar_text ="""ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي وهي بدايات الجبر, ومن المهم فهم كيف كانت هذه الفكرة الجديدة مهمة, فقد كانت خطوة نورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة, الجبر کان نظرية موحدة تتيح الأعداد الكسرية والأعداد اللا كسرية, والمقادير الهندسية وغيرها, أن تتعامل على أنها أجسام جبرية, وأعطت الرياضيات ككل مسارا جديدا للتطور بمفهوم أوسع بكثير من الذي كان موجودا من قبل, وقم وسيلة للتنمية في هذا الموضوع مستقبلا. وجانب آخر مهم لإدخال أفكار الجبر وهو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث من قبل"""
```

# tokenizing

# delete Stop Words

```python
from nltk.corpus import stopwords
print(stopwords.words("Arabic"))
```

```python
stop_words = set(stopwords.words('Arabic'))
words = words_tokniz
for word in words:
        if word not in stop_words:
                print(word)
```

# Tokenizing sentences


```python
from nltk.tokenize import sent_tokenize, word_tokenize
print(sent_tokenize(Ar_text))
```

# delete ponctuations


```python
stop_words = set(stopwords.words('Arabic'))
words = word_tokenize(Ar_text)
for word in words:
        if word not in string.punctuation:
                print(word)
```

# lemmatization

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
sentence_words = nltk.word_tokenize(Ar_text)
for word in sentence_words:
    if word in string.punctuation:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemmatization"))
print()
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```

# Stemming

```python
tokens = nltk.word_tokenize(Ar_text)

porter = nltk.PorterStemmer()
[porter.stem(t) for t in tokens]

lancaster = nltk.LancasterStemmer()
[lancaster.stem(t) for t in tokens]


wnl = nltk.WordNetLemmatizer()
[wnl.lemmatize(t) for t in tokens]
```
